## Data Analysis Question 1

The two most important customers of the firm may have been hacked. Verify if there are any fraudulent transactions in their history. For privacy reasons, you only know that their cardholder IDs are 2 and 18.

* Using hvPlot, create a line plot representing the time series of transactions over the course of the year for each cardholder separately. 

* Next, to better compare their patterns, create a single line plot that containins both card holders' trend data.  

* What difference do you observe between the consumption patterns? Does the difference suggest a fraudulent transaction? Explain your rationale in the markdown report.

```python
# loading data for card holder 2 and 18 from the database
# Write the query
query = """
        SELECT 
        card_holder.id,
        transaction.date, 
        transaction.amount, 
        transaction.card, 
        transaction.id_merchant, 
        merchant_category.name  

        FROM card_holder
        INNER JOIN credit_card ON card_holder.id = credit_card.cardholder_id
        INNER JOIN transaction ON transaction.card = credit_card.card
        INNER JOIN merchant ON transaction.id_merchant = merchant.id_merchant_category
        INNER JOIN merchant_category ON transaction.id_merchant = merchant_category.id
        WHERE card_holder.id in (2,18);
        """
# Create a DataFrame from the query result. HINT: Use pd.read_sql(query, engine)
card_holders_df = pd.read_sql(query, engine)

card_holders_df.head()
```

```python
# Create a dataframe for cardholder 2
card_holder2_df = card_holders_df[card_holders_df['id'] == 2].copy()

card_holder2_df = card_holder2_df.set_index('date')
card_holder2_df.index = pd.to_datetime(card_holder2_df.index)
card_holder2_df.head()
```

```python
# Plot for cardholder 2
card_holder2_df.hvplot.line(
    x="date",
    y="amount",
    xlabel="Date",
    ylabel="Amount",
    title="Transactions by Id Holder 2",
    rot = 90
)
```

```python
# Create a dataframe for cardholder 18
card_holder18_df = card_holders_df[card_holders_df['id'] == 18].copy()

card_holder18_df = card_holder18_df.set_index('date')

card_holder18_df.head()
```

```python
# Plot for cardholder 18
card_holder18_df.hvplot.line(
    x="date",
    y="amount",
    xlabel="Date",
    ylabel="Amount",
    title="Transactions by Id Holder 2",
    rot = 90
)
```

```python
# Combined plot for card holders 2 and 18
combined_plot = (
    card_holder2_df.hvplot.line(
        x="date",
        y="amount",
        xlabel="Date",
        ylabel="Amount",
        title="Transactions by Id Holder 2",
        rot=90
    ) *
    card_holder18_df.hvplot.line(
        x="date",
        y="amount",
        xlabel="Date",
        ylabel="Amount",
        title="Transactions by Id Holder 18",
        rot=90
    )
)

combined_plot
```
